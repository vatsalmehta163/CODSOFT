```python
print("Welcome to simple calculator !!")


def Calculator():
  while True:
    choice=int(input("what do you want to perform ? \n 1. Addition \n 2. Subtraction \n 3. Multiplication \n 4. Division \n "))
    Number_1=int(input("Enter the first number:"))
    Number_2=int(input("Enter the second number"))

    if choice==1:
      print(f"The result is: {Number_1+Number_2}")
      break
    elif choice==2:
      print(f"The result is: {Number_1-Number_2}")
      break
    elif choice==3:
      print(f"The result is: {Number_1*Number_2}")
      break
    elif choice==4:
      print(f"The result is: {Number_1/Number_2}")
      break
    else:
      print("Invalid choice! Please select a valid operation.")
def main():
  while True:
    Calculator()
    user_choice1 = input("Do you want to continue? Type 'yes' or 'no': ").strip().lower()
    if user_choice1 != 'yes':
      print("Thank you !")
      break
if __name__ == "__main__":
    main()

```
